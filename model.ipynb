```
from keras.models import load_model
new_model = load_model('/content/drive/MyDrive/portfolio/dogs_vs_cats/dogs_vs_cats.h5')
```

```
from PIL import Image
import numpy as np

# 予測できるように画像の配列を整備
im = Image.open('/content/drive/MyDrive/portfolio/dogs_vs_cats/cat.4821.jpg')
im = im.resize((150, 150))
im = np.expand_dims(im, axis=0)
im = np.array(im, np.float32)
im /= 255

#予測
result_predict = new_model.predict(im)
print(result_predict)
```

```
if result_predict < 0.50:
  print('予測結果は猫です。')
else:
  print('予測結果は犬です。')
```
